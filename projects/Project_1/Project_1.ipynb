```python
#get the random weights between -10 to 10
def get_random_weights(n):
  w = [0] * (n+1)
  for i in range(n+1):
    w[i] = random.randint(-10, 10)
  
  return w
```

```python
# get the product of vector x and w(weight)
def get_total_product(x, w):
  n = len(x) + 1 # added 1 for bias
  res = 0
  for i in range(n):
    if i == 0:
      res += w[i]
    else:
      res += (x[i - 1] * w[i])
  return res
```

```python
# return the sign of y_head = sign(W,X)
def predict(x, w):
  dot_product = get_total_product(x, w)
  # print("dot_product ", dot_product)
  return (1 if dot_product > 0 else -1)
```

```python
#update the weights
def update_weights(w, delta_w):
  n = len(w)
  new_w = [0] * n
  for i in range(n):
    new_w[i] = w[i] + delta_w[i]
  return new_w
```

```python
#calculate delta w and return the new delta
def calculate_delta_w(delta_w, alpha, x):
  n = len(delta_w)
  new_delta_w = [0] * n
  # print("delta_w_len", n)
  for i in range(n):
    if i == 0:
      new_delta_w[i] = alpha
    else:
      new_delta_w[i] = alpha * x[i - 1]
  # print("delta_w", delta_w)
  return new_delta_w
```

```python
#Train function
#data_train: data set, could be training set or test set
#letter_train_bitmap: a letter‚Äôs image represented as a one-dimensional bitmap of 0‚Äôs and 1‚Äôs 
#letter_train_val: the character corresponding to the letter shown in the image.
#
def train(letter_train_bitmap, letter_train_val, data_train=TRAINING_SET):
  y = [1 if letter[1] == letter_train_val else -1 for letter in data_train]
  w = get_random_weights(len(letter_train_bitmap))
  # print("random weights ", w)
  alpha = 0.01
  n = len(data_train)
  learning = True
  while learning:
    learning = False
    delta_w = [0] * len(w)
    for i in range(n):
      x_t = data_train[i][0]
      y_pred = predict(x_t, w)

      # update rule following p. 35 of Artificial Intelligence Engines. 
      # if y[i] < 0 and y_pred > 0:
      #   delta_w = calculate_delta_w(delta_w, -alpha, x_t)
      #   learning = True
      # if y[i] > 0 and y_pred < 0:
      #   delta_w = calculate_delta_w(delta_w, alpha, x_t)
      #   learning = True

      # applying the rule of equation 1.4 in the textbook
      error = y[i]-y_pred
      if (y[i] < 0 and y_pred > 0) or (y[i] > 0 and y_pred < 0):
        delta_w = calculate_delta_w(delta_w, alpha*error, x_t)
        learning = True

      w = update_weights(w, delta_w)

  # print("modified weights ", w)
  return w
```

# 5. Repeat experiment (3) for the rest of the letters, for a total of 26 trained perceptrons. If any perceptron fails to converge, implement the ‚Äúpocket algorithm‚Äù as described at the end of Section 1.2.1.1 of the textbook on p. 8.

## Note: for the experiment (3), our algorithm always converge. But, we also implement the pocket algorithm for learning and experiment.
## In brief, we have 3 algorithms:
- train(): train all perceptrons without pocket algorithm
- train_with_pocket_alg_2(): train all perceptrons with pocket algorithm. But runtime is O(N^2)
- train_with_pocket_alg_optimized(): train all perceptrons with pocket algorithm. But this algorithm is optimized based on professor's suggestion

```python
# Helper function to get number of classification errors
def get_num_misclassification(w, test_letter):
  res = []
  for letter_bitmap, letter_val in TRAINING_SET:
    res.append([predict(letter_bitmap, w), letter_val, test_letter])
  
  # print(res)
  misclassification = 0 # number of classification errors
  for ans, letter, test_letter in res:
    if (ans == 1 and letter != test_letter) or (ans == -1 and letter == test_letter):
      misclassification += 1
  return misclassification
```

```python
# following the suggestion from https://stats.stackexchange.com/questions/51200/pocket-algorithm-for-training-perceptrons
# Note: this algorithm takes a significant long time to train since it runs in O(n^2) run time
# Basically, we get number of misclassification on each images
def train_with_pocket_alg_2(letter_train_bitmap, 
                            letter_train_val, 
                            epochs=100, 
                            data_train=TRAINING_SET):
  y = [1 if letter[1] == letter_train_val else -1 for letter in data_train]
  w = get_random_weights(len(letter_train_bitmap))
  best_w = []
  best_error = get_num_misclassification(w, letter_train_val)
  alpha = 0.01
  n = len(data_train)
  for j in range(epochs):
    delta_w = [0] * len(w)
    for i in range(n):
      x_t = data_train[i][0]
      y_pred = predict(x_t, w)

      # applying the rule of equation 1.4 in the textbook
      error = y[i]-y_pred
      if (y[i] < 0 and y_pred > 0) or (y[i] > 0 and y_pred < 0):
        delta_w = calculate_delta_w(delta_w, alpha*error, x_t)

      w = update_weights(w, delta_w)
      curr_err = get_num_misclassification(w, letter_train_val)
      if curr_err < best_error:
        best_w = w
        best_error = curr_err
  return best_w
```

```python
# optimize train() function based on professor's suggestion
# Note: this algorithm runs faster than the previous algorithm since
# we only save the best weight after each epoch
def train_with_pocket_alg_optimized(letter_train_bitmap, 
                            letter_train_val, 
                            epochs=100, 
                            data_train=TRAINING_SET):
  y = [1 if letter[1] == letter_train_val else -1 for letter in data_train]
  w = get_random_weights(len(letter_train_bitmap))
  best_w = w
  alpha = 0.01
  n = len(data_train)
  for j in range(epochs):
    misclassification = 0
    delta_w = [0] * len(w)
    for i in range(n):
      x_t = data_train[i][0]
      y_pred = predict(x_t, w)

      # applying the rule of equation 1.4 in the textbook
      error = y[i]-y_pred
      if (y[i] < 0 and y_pred > 0) or (y[i] > 0 and y_pred < 0):
        delta_w = calculate_delta_w(delta_w, alpha*error, x_t)
        misclassification += 1

      w = update_weights(w, delta_w)
    if misclassification == 0:
      break
    else: # stop, otherwise save best_weights
      best_w = w
  return best_w
```
